```python
import matplotlib.pyplot as plt

plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
import numpy as np
np.random.seed(42)
```

```python
m = 100
X = 2 * np.random.rand(m, 1)
y = 4 + 3 * X
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
y2 = y + np.random.randn(m, 1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y2, "b.")
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
X_b = np.c_[np.ones((m, 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y2)
theta_best
```

```python
X_p = 2* np.random.rand(m, 1)
Y_p = theta_best[0] + theta_best[1] * X_p

plt.figure(figsize=(6, 4))
plt.plot(X_p, Y_p, "r-")
plt.plot(X, y2, "b.")
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y2)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

# Polynomial Regression

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
X_poly
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

```python
X_p = np.sort(6 * np.random.rand(m, 1) - 3, axis=None)
Y_p = lin_reg.intercept_[0] + lin_reg.coef_[0][0] * X_p + lin_reg.coef_[0][1] * X_p**2

plt.figure(figsize=(6, 4))
plt.plot(X_p, Y_p, "r-")
plt.plot(X, y, "b.")
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```
